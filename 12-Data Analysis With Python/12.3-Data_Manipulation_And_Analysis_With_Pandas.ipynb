```python
### Renaming the columns
df = df.rename(columns={'diagnosis':'disease'})
df.head()
```

```python
### Change datatypes 

df['new_value'] = df['x.radius_mean'].astype(float)
```

```python
df.head()
```

```python
### Apply the functions

df['Newvalue'] = df['new_value'].apply(lambda x:x*2)
df.head()
```

```python
## Data Aggregation and grouped data
grouped_mean = df.groupby('disease')['Newvalue'].mean()
grouped_mean
```

```python
### Aggregate multiple functions

grouped_agg = df.groupby('your column')['value'].agg(['mean', 'sum', 'count'])
```

## Merging and joining DataFrames

```python
df1 = pd.DataFrame({'key':['A', 'B', 'C'], 'value1':[1,2,3]})
df2 = pd.DataFrame({'key':['A', 'B', 'D'], 'value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge DataFrame on the 'key column'
pd.merge(df1, df2, on="key", how="inner")  ### Only
```

```python
pd.merge(df1, df2, on="key", how="outer")
```

```python
pd.merge(df1, df2, on="key", how="left")
```

```python
pd.merge(df1, df2, on="key", how="right")
```
